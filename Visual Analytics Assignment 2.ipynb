```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    tokens = word_tokenize(text)  # Tokenize
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords    
    return tokens

# # Example text
# data = ["Text analysis is an interesting field.", 
#         "NLP helps in extracting themes from large datasets."]

# Preprocess the text
preprocessed_data = [preprocess_text(text) for text in df['text']]
print(preprocessed_data)
```

```python
cleaned_texts = []
emp = []
for i in preprocessed_data[:]:
    emp= []
    for word in i:
        word = word.strip('_')
        if word.startswith('http') or len(word) == 1 or len(word) == 2 or \
        word.lower() in ['trump', 'hillary', 'donaldtrump', 'donald', 
                         'hillaryclinton', 'clinton', 'realdonaldtrump', 'thank', 'amp',
                        'trumps', 'president', 'vote']:
            continue
        emp.append(word)
    cleaned_texts.append(' '.join(emp))
print(cleaned_texts)
```

```python
df['cleaned_text'] = cleaned_texts
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['cleaned_text'])
tfidf_matrix_array = tfidf_matrix.toarray()

# Get the feature names and the TF-IDF scores
feature_names = vectorizer.get_feature_names_out()
feature_names
```

```python
tfidf_matrix_array.shape
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=15, random_state=42)  # Number of topics -- had use 10,15,20
lda.fit(tfidf_matrix_array)

# Print the top words for each topic
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))

n_top_words = 10
print_top_words(lda, feature_names, n_top_words)
```

```python
# categorize topic for each text
topic_distribution = lda.transform(tfidf_matrix_array)
dominant_topics = topic_distribution.argmax(axis=1)
# dominant_topics
df['dominant_topic_15'] = dominant_topics
```
